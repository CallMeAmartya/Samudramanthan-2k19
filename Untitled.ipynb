```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv("train__updated.csv")
```

```python
df.head()
```

```python
df.columns.tolist()
```

**Data Visualisation**

```python
missing_df = df.isnull().sum(axis=0).reset_index()
missing_df.columns = ['column_name', 'missing_count']
missing_df = missing_df.loc[missing_df['missing_count']>0]
missing_df = missing_df.sort_values(by='missing_count')

ind = np.arange(missing_df.shape[0])
width = 0.9
fig, ax = plt.subplots(figsize=(12,18))
rects = ax.barh(ind, missing_df.missing_count.values, color='#34495e')
ax.set_yticks(ind)
ax.set_yticklabels(missing_df.column_name.values, rotation='horizontal')
ax.set_xlabel("Count of missing values")
ax.set_title("Number of missing values in each column")
plt.show()
```

```python
df.fillna(method='ffill', inplace=True)
df.fillna(method='bfill', inplace=True)
df.isnull().sum()
```

```python
s=df.std()==0
m=[]
```

```python
for each in range(len(s)):
    if s[each]==True:
        m.append(s.index[each])
```

```python
m
```

```python
df['wbt_indicator'].describe() 
```

**Normalisation**

```python
normalized_df=df.drop(['index', 'wave_height', 'source_exclusion_flags', 'dpt_indicator'], axis=1)
normalized_df=normalized_df.drop(m, axis=1)
normalized_df=(normalized_df-normalized_df.mean())/normalized_df.std()
```

```python
normalized_df.head()
```

```python
Xn = normalized_df.values
yn = df['wave_height'].values

from sklearn.model_selection import train_test_split

Xn_train, Xn_val, yn_train, yn_val = train_test_split(Xn,yn,test_size=0.25)
```

**MLP**

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import validation_curve
MLP = MLPClassifier(alpha=0.005, max_iter=500)
MLP.fit(Xn_train, yn_train)
yn_pred = MLP.predict(Xn_val)

from sklearn import metrics
print('RMSE:', np.sqrt(metrics.mean_squared_error(yn_val, yn_pred)))
```

**RF**

```python
from sklearn.ensemble import RandomForestClassifier
RF = RandomForestClassifier(n_estimators=500)
RF.fit(Xn_train, yn_train)
yn_pred = RF.predict(Xn_val)

from sklearn import metrics
print('RMSE:', np.sqrt(metrics.mean_squared_error(yn_val, yn_pred)))
```

**XGB Regressor**

```python
from xgboost import XGBRegressor
XG = XGBRegressor()
XG.fit(Xn_train, yn_train)
yn_pred = XG.predict(Xn_val)

from sklearn import metrics
print('RMSE:', np.sqrt(metrics.mean_squared_error(yn_val, yn_pred)))
```

**LR without regularisation**

```python
from sklearn.cross_validation import train_test_split
t=df.drop(['index', 'wave_height', 'source_exclusion_flags', 'dpt_indicator'], axis=1)
t=df.drop(m, axis=1)
X=t.values
y=df['wave_height'].values
X_train, X_test, y_train, y_test=train_test_split(X,y)
```

```python
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(Xn_train,yn_train)
yn_pred = linreg.predict(Xn_val)

from sklearn import metrics
print('RMSE:', np.sqrt(metrics.mean_squared_error(yn_val, yn_pred)))
```
